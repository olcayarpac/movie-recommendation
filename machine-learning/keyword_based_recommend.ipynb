```python
# STEMMERS FOR FINDING THE PLAIN VERSION OF WORD
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
```

```python
porter = PorterStemmer()
lancaster=LancasterStemmer()
```

```python
#proide a word to be stemmed
wortto = 'terrorism'
print("Porter Stemmer")
print(porter.stem(wortto))
print("Lancaster Stemmer")
print(lancaster.stem(wortto))
```

```python
# STEM KEYWORDS LIST AND CREATE NEW ONE
porterKeywords = []
lancasterKeywords = []
```

```python
punctuations = str.maketrans('', '', string.punctuation)
stop_words = [word.translate(punctuations) for word in stopwords.words('english')]
for keyword in keywordList:
    portered = porter.stem(keyword)
    lancastered = lancaster.stem(keyword)
    porterKeywords.append(portered)
    lancasterKeywords.append(lancastered)
stop_words.append('cant')
```

```python
print(stop_words)
```

```python
# START TO EXTRACT KEYWORDS
```

```python

for i, row in movies.iterrows():
    keywords = []
    for word in row['description'].translate(punctuations).lower().split():
        if(word in stop_words):
            continue
            
        pStemmed = (porter.stem(word) if not porter.stem(word) in stop_words else '-')
        lStemmed = (lancaster.stem(word) if not lancaster.stem(word) in stop_words else '-')
        
        if (word in keywordList):
            keywords.append(word)
        if not (pStemmed in stop_words):
            if (pStemmed in keywordList):
                keywords.append(pStemmed)
            elif (pStemmed in porterKeywords):
                keywords.append(keywordList[porterKeywords.index(pStemmed)])
            
        if not (lStemmed in stop_words):
            if (lStemmed in keywordList):
                keywords.append(lStemmed)
            elif (lStemmed in lancasterKeywords):
                keywords.append(keywordList[lancasterKeywords.index(lStemmed)])
    for genre in row['genre'].split(','):
        keywords.append(genre.lower())
    print(keywords)
    movies.at[i, 'keywords'] = ' '.join(str(keyword) for keyword in list(set(keywords)))
```

```python
# CREATE RECOMMENDATION SYSTEM AND FUNCTIONS
```

```python
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
```

```python
tfidf_keywords = tf.fit_transform(movies['keywords'])
```

```python
cosine_sim_keywords = linear_kernel(tfidf_keywords, tfidf_keywords)
```

```python
indices = pd.Series(movies.index, index=movies['imdb_title_id'])
```

```python
# Function that recommends based on keywords similarity
# movie id parameters will be without 'tt', only integer part
def get_recommendations_keywords(movieId):
    idx = indices[("tt" + str(movieId).zfill(7))]
    sim_scores = list(enumerate(cosine_sim_keywords[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    movie_indices = [i[0] for i in sim_scores]
    return [movies.at[indice, 'imdb_title_id'] for indice in movie_indices]
```
